```python
sns.set(style="whitegrid")

# Line Plot
plt.figure(figsize=(12, 6))
for column in Agriculture_data.columns[1:-1]:
    plt.plot(Agriculture_data["Year"], Agriculture_data[column], label=column)
plt.plot(Agriculture_data["Year"], Agriculture_data["Gross Total"], label="Gross Total", linewidth=2.5, color="black")
plt.title("U.S. Agricultural Emissions (1990-2022)")
plt.xlabel("Year")
plt.ylabel("Emissions (MMT CO2 eq.)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Stacked Area Plot
plt.figure(figsize=(12, 6))
plt.stackplot(
    Agriculture_data["Year"],
    Agriculture_data["Crop Cultivation"],
    Agriculture_data["Livestock"],
    Agriculture_data["Fuel Combustion"],
    labels=["Crop Cultivation", "Livestock", "Fuel Combustion"],
    colors=["#ff6666", "#3399ff", "#66ff66"]
   # colors=["#ff9999", "#66b3ff", "#99ff99"]
)
plt.title("Contribution to Agricultural Emissions Over Time")
plt.xlabel("Year")
plt.ylabel("Emissions (MMT CO2 eq.)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
bar_width = 0.6
x = Agriculture_data["Year"]

plt.bar(x, Agriculture_data["Crop Cultivation"], label="Crop Cultivation", color="#ff6666", width=bar_width)
plt.bar(x, Agriculture_data["Livestock"], bottom=Agriculture_data["Crop Cultivation"], label="Livestock", color="#3399ff", width=bar_width)
plt.bar(
    x,
    Agriculture_data["Fuel Combustion"],
    bottom=Agriculture_data["Crop Cultivation"] + Agriculture_data["Livestock"],
    label="Fuel Combustion",
    color="#66ff66",
    width=bar_width,
)

# Labels and legend
plt.title("Agricultural Emissions by Source (1990-1994)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Emissions (MMT CO2 eq.)", fontsize=12)
plt.xticks(x)
plt.legend()
plt.tight_layout()
plt.show()
```
