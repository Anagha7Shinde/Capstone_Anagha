```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=new_data, palette='tab10', linewidth=2)
plt.title('U.S. Agricultural Sector Emissions Over Time (MMT CO2 eq.)', fontsize=16)
plt.xlabel('Year', fontsize=8)
plt.ylabel('Emissions (MMT CO2 eq.)', fontsize=12)
plt.legend(title='Sector', loc='upper right')

# Rotate only the y-axis labels
plt.xticks(rotation=45)  # Change 45 to any angle you prefer

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.stackplot(new_data['U.S. Emissions, Agricultural Sector, MMT CO2 eq.'], new_data['Crop cultivation'], new_data['Livestock'], new_data['Fuel combustion'], labels=['Crop cultivation', 'Livestock', 'Fuel combustion'], alpha=0.6)

plt.title('U.S. Emissions from Agricultural Sector (MMT CO2 eq.)')
plt.xlabel('Year')
plt.ylabel('Emissions (MMT CO2 eq.)')
plt.legend(loc='upper left')


plt.tight_layout()
plt.show()
```

```python
 Greenhouse_data_by_state = pd.read_csv('../data/Greenhouse_Gas.csv')

Greenhouse_data_by_state.head()
```

```python
Greenhouse_data_by_state.rename(columns={'Greenhouse Gas Gross Total, MMT CO2 eq.': 'State'}, inplace=True)
```

```python
col_list=Greenhouse_data_by_state.columns.to_list()[1:]
```

```python
Greenhouse_data_by_state.melt(id_vars='State', value_vars=col_list)
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=Greenhouse_data_by_state, x='variable', y='value', hue='State', marker='o')
plt.title('State-wise Value Over Time')
plt.xlabel('variable')
plt.ylabel('value')
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
